[Exploratory Data Analysis (EDA) is an important part of any data analysis and is a process of examining or understanding the data and extracting insights or main characteristics of the data. It refers to the critical process of performing initial investigations on data so as to discover patterns,to spot anomalies,to test hypothesis and to check assumptions with the help of summary statistics and graphical representations. Originally developed by American mathematician John Tukey in the 1970s, EDA techniques continue to be a widely used method in the data discovery process today.]

#### Visualizing the data using various plots to understand the various characteristics of the data

```python
# detection of the various outliers in the data
df.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# observing ship mode
sns.pairplot(df, hue = 'Ship Mode')
```

```python
sns.countplot(x = df['Ship Mode'])
```

```python
# observing region
sns.pairplot(df, hue = 'Region')
```

```python
sns.countplot(x = df['Region'])
```

```python
# observing category
sns.pairplot(df,hue='Category')
```

```python
sns.countplot(x = df['Category'])
```

Observation: The Standard Class, Office Supplies Category and West Region has the highest count in the Ship Mode, Category and the Region respectively.

```python
# visualizing the correlation of features using a correlation matrix
f,ax = plt.subplots(figsize =(15,10))
sns.heatmap(df.corr(method='pearson'), annot=True, cmap="inferno", lw =6, linecolor='black')
```

```python
# creating a pie chart in order to observe the sales of different sub-categories
plt.figure(figsize=(10,10))
df['Sub-Category'].value_counts().plot.pie(autopct= '%3.1f%%')
plt.show()
```

```python
# creating a scatter plot to observe the spread of datapoints between sales and profits
fig, ax = plt.subplots(figsize =(10,6))
ax.scatter(df["Sales"] , df["Profit"], color = 'blue')
ax.set_xlabel('Sales in USD$')
ax.set_ylabel('Profit/Loss in USD$')
plt.title('Sales and Profit', fontsize = 20)
plt.show()
```

```python
# creating a line plot for profit vs quantity
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df, color="red")
plt.legend()
plt.show()
```

```python
# total Sales for each State
df_state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending = False).plot.bar(figsize = (15, 5), 
                                                                                            color = 'purple')
plt.ylabel('Total Sales in USD')
plt.xlabel('States')
plt.title('Total State-Wise Sales', fontsize = 15)
plt.show()
```

Observation: California, New York and Texas are the top 3 states having the highest sales whereas Maine, West Virginia and North Dakota have the lowest sales.

```python
# total Profit/Loss for each State
df_state_profit = df.groupby('State')['Profit'].sum().sort_values(ascending = False).plot.bar(figsize = (15, 5), 
                                                                                              color = 'purple')
plt.ylabel('Total Profit/Loss in USD')
plt.xlabel('States')
plt.title('Profit/Loss across various states', fontsize = 15)
plt.show()
```

Observation: The states of California and New York have the highest profits and they are nearly same whereas Texas suffers the highest loss among all the states.

```python
# average discount for each State
df_state_profit = df.groupby('State')['Discount'].mean().sort_values(ascending = False).plot.bar(figsize = (15, 5), 
                                                                                                 color = 'purple')
plt.ylabel('Average Discount')
plt.xlabel('States')
plt.title('Discounts given in each State', fontsize = 15)
plt.show()
```

Observation: The states of Illinois and Texas offers the highest discount rates whereas New York and California which had the maximum profit offers considerably low discount rates and there are many states which offer no discount at all.

```python
# total Profit/loss according to the different sub-categories
df_category_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending = False).plot.bar(figsize = (15, 10), 
                                                                                              color = 'purple')
plt.ylabel('Total Profit/Loss in USD')
plt.xlabel('Sub-categories')
plt.title('Sub-category wise Profit/Loss', fontsize = 15)
plt.show()
```

Observation: The sub-categories having the maximum profit are copies, after which phones and accessories have nearly the same profit whereas selling tables makes for the highest loss.

```python
# profit/loss by discount level
plt.figure(figsize = (10, 5))
sns.lineplot(x=df['Discount'], y=df['Profit'], color = 'Red', label = 'Discount Level')
plt.ylabel('Profit/Loss in USD') 
plt.xlabel('Discount Level')
plt.title('Profit/Loss by Discount Level', fontsize = 15)
plt.show()
```

Observation: The worst losses occur approximately at a discount of approximately 50%. When the discount is between 0% and 20%, profits occur but beyond that margin, losses occur.
